```python
import os
import matplotlib.pyplot as plt
from PIL import Image

# Path to the folder containing images
images_folder = "/Users/apple/Desktop/MajorProject/DataSet/test/TURBERCULOSIS"

# Get a list of all images in the folder
image_files = sorted([os.path.join(images_folder, file) for file in os.listdir(images_folder)])

# Create a figure object with a 2x4 grid of subplots
fig, axs = plt.subplots(nrows=2, ncols=4, figsize=(15, 8))

# Iterate over the first 8 images and display them in the subplots
for i, ax in enumerate(axs.flat):
    if i < 8:
        # Load image
        image = Image.open(image_files[i])

        # Display image
        ax.imshow(image, cmap='gray')
        ax.set_title(f"Tuberculosis X-ray {i+1}")
        ax.axis('off')
        
plt.tight_layout()
plt.show()
```
